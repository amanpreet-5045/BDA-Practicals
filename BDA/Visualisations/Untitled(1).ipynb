```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("sample_dataset.csv")

# Display the first few rows of the dataset
print(df.head())
```

```python
# Plot a histogram of ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), bins=10, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot a boxplot of salary by city
plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='Salary', data=df)
plt.title('Salary Distribution by City')
plt.xlabel('City')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a bar chart of count of individuals by city
plt.figure(figsize=(10, 6))
sns.countplot(x='City', data=df, palette='viridis')
plt.title('Count of Individuals by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
